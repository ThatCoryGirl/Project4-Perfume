```python
import pandas as pd
import sqlite3

# Step 1: Read CSV file into a DataFrame
csv_file = 'C:/Users/Amanda/Desktop/Analytics/PRoject 4/Resources/perfume_database.csv'
df = pd.read_csv(csv_file)

# Step 2: Connect to SQLite Database
# Specify the desired path for the SQLite database file
sqlite_file = 'C:/Users/Amanda/Desktop/Analytics/PRoject 4/Resources/perfume.db'

conn = sqlite3.connect(sqlite_file)
cur = conn.cursor()

# Step 3: Create Table
table_name = 'No_Reviews'
df.to_sql(table_name, conn, index=False, if_exists='replace')  # Replace if table exists

# Commit changes and close connection
conn.commit()
conn.close()
```
